```python
def year_list (raceidj,raceidt, uresults):
    jerusalem = dict()
    tiberias = dict()
    for uresult in uresults:
        if (uresult["raceid"] == raceidj) and (uresult["Meters"] == '42195'):
            jerusalem[uresult["PKey"]] =(int(uresult['PSECResults']) , int(uresult["PAge"]))
        
        #print(uresult["Date"])
        elif uresult["raceid"] == raceidt and (uresult["Meters"] == '42195'):
            tiberias[uresult["PKey"]] =(int(uresult['PSECResults']) , int(uresult["PAge"]))
    return(jerusalem ,tiberias)
```

```python
jerusalem16, tiberias16 = year_list("17245",'16751',uresults )
jerusalem17, tiberias17 = year_list('28506','21719',uresults )
jerusalem18, tiberias18 = year_list('28495','27481',uresults )
```

```python
def gap(jerusalem, tiberias):
    gapx = list()
    gapy = list()
    for jer in jerusalem.keys():
        if jer in tiberias.keys():
        #if type(gap[(tiberias16[jer][1] - tiberias16[jer][1]%10 )]) != list:
        #    agestack = list(gap[(tiberias16[jer][1] - tiberias16[jer][1]%10 )])
            gapx.append(tiberias[jer][1] - tiberias[jer][1]%10 )
            gapy.append(jerusalem[jer][0] - tiberias[jer][0] )
    return(gapx,gapy)
print(gap(jerusalem16, tiberias16))
```
